```python
gender_submission=pd.read_csv('/kaggle/input/titanic/gender_submission.csv')
test_data=pd.read_csv('/kaggle/input/titanic/test.csv')
train_data=pd.read_csv('/kaggle/input/titanic/train.csv')
```

```python
print(train_data)
```

```python
train_df=pd.DataFrame(train_data)
test_df=pd.DataFrame(test_data)
gender_submission_df=pd.DataFrame(gender_submission)
```

```python
# train_df.dropna(inplace=True)
X_train=train_df.drop(['PassengerId','Survived','Name','Ticket','Fare','Cabin','SibSp','Parch','Embarked'],axis=1);
X_train['Sex'].replace(['male','female'],[1,0],inplace=True)
y_train=train_df['Survived']

print(X_train)
```

```python
my_imputer = SimpleImputer()
X_train= my_imputer.fit_transform(X_train)
```

```python
X_test=test_df.drop(['PassengerId','Name','Ticket','Fare','Cabin','Parch','SibSp','Embarked'],axis=1);
X_test['Sex'].replace(['male','female'],[1,0],inplace=True)
X_test=my_imputer.fit_transform(X_test)
y_test=gender_submission_df['Survived']
```

```python
print(y_train)
```

```python
# logistic_model = LogisticRegression()
# logistic_model.fit(X_train, y_train)
# classifier.fit(X_train, y_train)
model=Sequential([
    Dense(units=200,activation='relu'),
    Dense(units=150,activation='relu'),
    Dense(units=100, activation='relu'),
    Dense(units=50, activation='relu'),
    Dense(units=15, activation='relu'),
    Dense(units=1,activation='sigmoid')
])
model.compile(loss=BinaryCrossentropy())
model.fit(X_train,y_train, epochs=200)
```

```python
# y_pred = logistic_model.predict(X_test)
# print(y_pred)
prediction=model.predict(X_test)
prediction = prediction.astype(float)
for i in range(418):
    if prediction[i]>=0.5:
        prediction[i]=1
    else:
        prediction[i]=0
```
